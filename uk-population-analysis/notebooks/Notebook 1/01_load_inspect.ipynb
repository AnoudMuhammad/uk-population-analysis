# Notebook 1: LOad and Inspect

## Purpose 
Load the ONS UK Population Estimates file (1838-2024) and understand its structure before any cleaning takes place.

## Source 
Office for National Statistics (ONS) - Mid-Year Population Estimates
Pulished: 26 September 2025
File: ukpopulationestimates18382024.xlsx

```python
import pandas as pd
import openpyxl

RAW_FILE = 'data/raw/ukpopulationestimates18382024.xlsx'

wb = openpyxl.load_workbook(RAW_FILE, read_only=True)
print("All sheets in this file.:")
for i, name in enumerate(wb.sheetnames):
    print(f" {i+1}. {name}")
```

```python
df_t1 = pd.read_excel(RAW_FILE, sheet_name='Table 1', header=1)
print("Table 1 shape:", df_t1.shape)
print("\nFirst 5 rows:")
print(df_t1.head())
print("\nColumn names:", df_t1.columns.tolist())
print("nData types:")
print(df_t1.dtypes)
```

```python
df_t3 = pd.read_excel(RAW_FILE, sheet_name='Table 3', header=1)
print("Table 3 shape:", df_t3.shape)
print("\nFirst 8 rows:")
print(df_t3.iloc[:8,:6])
print("\nSex values in dataset:", df_t3['Sex'].unique())
print("\nAge values (first 15):", df_t3['Age'].head(15).tolist())
print("\nMissing values per column:")
print(df_t3.isnull().sum().sum(), "total missing cells")
```

## Inspection Findings

- **File contains 18 sheets**: Cover, Contents, Notes, Tables 1-18
- **Table 1**: UK total population 1851-2024. Two columns: Year, Total Population
- **Table 3**: UK population by sex + single year of age, 1971-2024
- Shape: 276 rows x 56 columns
- Ages: 0 to 89 (age 90 = top-coded)
- Sex: Persons / Males / Females
- Years: Mid-1971 to Mid-2024 (54 year columns)
- Format: WIDE - each year is a seperate columns
- ** Key challenge**: Must reshape Table 3 from WIDE to LONG format (one row per Age / Sex / Year combination) before analysis
